```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
import re

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('wordnet')

# Function to clean and preprocess text
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()

    # Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenization by whitespace
    tokens = text.split()

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Rejoin tokens into a single string
    text = ' '.join(tokens)
    return text

# Selecting relevant columns for preprocessing
columns_to_preprocess = ['Topic Resources', 'Symptoms and Signs', 'Diagnosis', 'Key Points', 'More Information']

# Preprocess each selected column
for column in columns_to_preprocess:
    data[column] = data[column].astype(str).apply(preprocess_text)

# Show the processed data
data[columns_to_preprocess].head()
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

# Ensure NLTK resources are available
nltk.download('stopwords')
nltk.download('wordnet')

# Load the dataset
data = pd.read_excel(file_path)

# Function to clean and preprocess text for all columns
def preprocess_text(text):
    if pd.isna(text):
        return ""
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenization by whitespace
    tokens = text.split()
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Rejoin tokens into a single string
    return ' '.join(tokens)

# Apply preprocessing to all text columns
text_columns = data.select_dtypes(include=['object']).columns  # This selects columns that contain text
data[text_columns] = data[text_columns].applymap(preprocess_text)

# Save the cleaned data to a new Excel file
data.to_excel('path_to_your_cleaned_file.xlsx', index=False)

print("Data preprocessing complete and saved to 'path_to_your_cleaned_file.xlsx'")
```

```python
import os
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import faiss

# List of uploaded files
files = [
    "/content/ATCDPP.csv", "/content/VOSDCI.csv", "/content/Gal.csv",
    "/content/Ggr_Link.csv", "/content/Hyr.csv", "/content/Ir.csv",
    "/content/MP.csv", "/content/MPP.csv", "/content/Sam.csv", "/content/Stof.csv"
]

# Function to preprocess text
def preprocess_text(text):
    text = text.lower()  # Lowercase text
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove non-alphabetic characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Load all files into a single DataFrame
all_data = pd.DataFrame()
for file in files:
    df = pd.read_csv(file)
    df = df.applymap(lambda x: preprocess_text(str(x)))  # Apply preprocessing
    all_data = pd.concat([all_data, df], ignore_index=True)

# Display the first few rows of the combined preprocessed data
all_data.head()
```
